<b>The Correlation Map (or Heatmap)</b> is a visual tool used to represent the strength and direction of the correlation between two or more variables. Here’s how to interpret it:
    
<b>Values in the Matrix:</b>

Each cell in the matrix shows the correlation coefficient between two variables. The correlation coefficient ranges from -1 to 1.

<b>1</b> indicates a perfect positive correlation (when one variable increases, the other also increases).
<b>-1</b> indicates a perfect negative correlation (when one variable increases, the other decreases).
<b>0</b> indicates no correlation (the variables have no linear relationship). 
<b>-0</b> means exactly the same as 0 — no relationship.

<b>Colors:</b>

The colors represent the strength of the correlation, usually following a gradient.

<b>Interpretation:</b>

Strong Positive Correlation (values close to 1): When one variable increases, the other also tends to increase. For example, in sales data, there may be a strong positive correlation between advertising spending and revenue.
Strong Negative Correlation (values close to -1): When one variable increases, the other tends to decrease. For example, in health data, there may be a strong negative correlation between exercise quantity and body weight.
Weak or No Correlation (values close to 0): The variables have no clear linear relationship. One variable does not reliably predict the other.

```python
class detection:

    @staticmethod

    def outlier_detect(ecom,col_name):

        # check your column is exists or not in your dataframe

        if col_name in ecom.columns:

            #using outlier remove by manually method

            column_data = ecom[col_name]


            Q1 = ecom[col_name].quantile(25/100)
            Q2 = ecom[col_name].quantile(50/100)
            Q3 = ecom[col_name].quantile(75/100)

            # counting Initial Quartile Range
            IQR = Q3-Q1

            lower_limit = Q1 - (1.5*IQR)
            upper_limit = Q3 + (1.5*IQR)

            a  = [i for i in ecom[col_name] if lower_limit > i or upper_limit < i]

            # here now check above process outlier in existing this perticuler column in dataset ?

            if a :
                #print("~> Outliers : ", a)
                print("~> Count of Outliers : ", len(a))
                print()
                print("~> Percentage of Outlier in your selectd column : " ,round((len(a)/len(ecom[col_name]))*100,2), "%")

                # represent by boxplot graph formate
                sns.boxplot(ecom[col_name])
                plt.show()

            else:
                print("Outlier are not found for your selected column in this dataset.")

        else:
            print(f"column {col_name} is does not exist in the csv file.")

```

```python
class clean1 :

    @staticmethod

    def trim_process(df,col_name):

        # check your column is exists or not in your dataframe

        if col_name in df.columns:

            column_data = df[col_name]

            #using outlier remove by manually method

            Q1 = df[col_name].quantile(25/100)
            Q2 = df[col_name].quantile(50/100)
            Q3 = df[col_name].quantile(75/100)

            # counting Initial Quartile Range
            IQR = Q3-Q1

            lower_limit = Q1 - (1.5*IQR)
            upper_limit = Q3 + (1.5*IQR)

            data1 = df[(df[col_name]>=lower_limit)  & (df[col_name]<=upper_limit)]


            # showing result from calculate by manually method
            print("~> Data before cleaning outliers within the data : " ,len(df))
            print("~> Number of outliers within the data            : " , len(df) - len(data1))
            print("~> Data after cleaning outliers within the data  : ", len(data1))

            # represent by boxplot graph formate
            sns.boxplot(x = col_name , data = data1)

            new_df = pd.DataFrame(data1)
            return new_df


        else:
            print(f"column '{col_name}' does not exist in this csv file.")
```

Here this is based on market and customer purchase scenario to I will make RFM analysis technique.

RFM stands for Recency – Frequency – Monetary value for following definitions.

<b>Recency</b> : Given a current or specific date in the past, Recency captures the last time that the customer made a transaction. inshort measures the time since the customer’s last transaction.

<b>Frequency:</b> Given a specific time window, Frequency captures the number of transactions the customer did during that window. inshort counts how many transactions a customer made within a specific time frame.

<b>Monetary Value or Revenue </b>: Given a specific window, Revenue captures the amount spent by the customer. inshort calculates the total spending by the customer within a specific time frame.

<b>Customer Segments:</b>

<b>Low Value Segment: </b> Customers who are less active, infrequent, and low spenders.

<b>Mid Value Segment: </b> Customers with moderate activity, spending, or inconsistent behavior.

<b>High Value Segment: </b> Customers who are highly active, frequent, and have a high spending profile.

<b>Approach:</b>

Calculate RFM Scores: Compute individual scores for Recency, Frequency, and Monetary Value to understand customer behavior.

Plot the RFM Values: Visualize the scores to analyze their ranges and distribution.

Clustering with K-Means: Apply the K-Means clustering algorithm to classify customers into segments (low, mid, high value).

Optimize Clusters: Use Silhouette and Inertia scores to determine the optimal number of clusters.

<b>Goal:</b>

To segment customers into actionable groups (e.g., low, mid, and high value) for personalized marketing and strategic business decisions.

```python
import pandas as pd

def order_cluster(cluster_field_name, target_field_name, df, ascending):
    # Create a new column for reordered cluster labels
    new_cluster_field_name = 'new_' + cluster_field_name
    
    # Group by cluster and calculate the mean of the target field
    cluster_means = df.groupby(cluster_field_name)[target_field_name].mean().reset_index()
    
    # Sort clusters based on the mean of the target field
    cluster_means = cluster_means.sort_values(by=target_field_name, ascending=ascending).reset_index(drop=True)
    
    # Add an index column to represent the new order of clusters
    cluster_means['index'] = cluster_means.index
    
    # Merge the original DataFrame with the new cluster means
    df_with_new_clusters = pd.merge(df, cluster_means[[cluster_field_name, 'index']], on=cluster_field_name)
    
    # Drop the old cluster field and rename the 'index' column to the original cluster field name
    df_with_new_clusters = df_with_new_clusters.drop([cluster_field_name], axis=1)
    df_with_new_clusters = df_with_new_clusters.rename(columns={"index": cluster_field_name})
    
    return df_with_new_clusters

# Applying the function to reorder clusters in the 'RecencyCluster' column based on 'Recency'
re_user = order_cluster('RecencyCluster', 'Recency', re_user, False)
```

we segmented our customers and found out who are the best ones. Now it’s time to measure one of the most important metric we should closely track: Customer Lifetime Value.

We invest in customers (acquisition costs, offline ads, promotions, discounts & etc.) to generate revenue and be profitable. Naturally, these actions make some customers super valuable in terms of lifetime value but there are always some customers who pull down the profitability. We need to identify these behavior patterns, segment customers and act accordingly.

Calculating Lifetime Value is the easy part. First we need to select a time window. It can be anything like 3, 6, 12, 24 months. By the equation below, we can have Lifetime Value for each customer in that specific time window:

LifeTimeValue = Total Gross Revenue - Total Cost

This equation now gives us the historical lifetime value. If we see some customers having very high negative lifetime value historically, it could be too late to take an action. At this point, we need to predict the future with machine learning: We are going to build a simple machine learning model that predicts our customers lifetime value.

Deciding the time frame really depends on your industry, business model, strategy and more. For some industries, 1 year is a very long period while for the others it is very short. In our example, we will go ahead with 6 months.

Since our feature set is ready, let’s calculate 6 months LTV for each customer which we are going to use for training our model.

```python
# Function to choose the value of k

from sklearn.neighbors import KNeighborsClassifier
def choose_k(X_train, y_train, X_test, y_test, max_k=10):
    k_values = list(range(1, max_k+1))
    accuracy_values = []
    
    for k in k_values:
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train, y_train)
        y_pred = knn.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)
        accuracy_values.append(accuracy)
    
    # Visualize the accuracy for different k values
    plt.figure(figsize=(10, 6))
    plt.plot(k_values, accuracy_values, marker='o', linestyle='-')
    plt.title('Accuracy vs. Number of Neighbors (k)')
    plt.xlabel('Number of Neighbors (k)')
    plt.ylabel('Accuracy')
    plt.xticks(k_values)
    plt.grid(True)
    plt.show()

    best_k = k_values[accuracy_values.index(max(accuracy_values))]
    return best_k
```

```python
# Choose the best k value
best_k = choose_k(X_train, y_train, X_test, y_test)
print("Best k:", best_k)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, f1_score

# Initialize model
models = {
    "LogisticRegression":LogisticRegression(),
    "KNeighborsClassifier" : KNeighborsClassifier(),
    "SVC" : SVC(),
    "DecisionTreeClassifier" : DecisionTreeClassifier(),
    "RandomForestClassifier" : RandomForestClassifier(),
    "GradientBoostingClassifier" : GradientBoostingClassifier(),
    "AdaBoostClassifier" : AdaBoostClassifier(),
    "XGBClassifier" : XGBClassifier(),
    "GaussianNB" : GaussianNB(),
    "MLPClassifier" : MLPClassifier()
}

results = {}

# Train each model and calculated metrics

for name, model in models.items():
    try:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred, average= "weighted")
        results[name] = {"Accuracy": accuracy, "F1-Score" : f1}

    except Exception as e:
        print(f"Error with moel {name} : {e}")

# find best model on F1-score
best_model = max(results, key = lambda x:results[x]["F1-Score"])
print(f"Best Model : , {best_model}")
print(F"Metrices : , {results[best_model]}")

# display all results
print()
print("Display all results")
print()
for model_name, metrics in results.items():
    print(f"{model_name} : {metrics}")
```
